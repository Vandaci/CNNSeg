```python
import sys 
sys.path.append('../src')
import cnnsegmentation as seg
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
proto_path='../model/deploy.prototxt'
caffe_model_path='../model/deploy.caffemodel'
test_pcd_path='../data/test.pcd'
test_cnnseg=seg.CNNSegmention()
test_cnnseg.forward(proto_path,caffe_model_path,test_pcd_path)
```

```python
instance_pt=test_cnnseg.outblobs['instance_pt']
print(instance_pt.shape)
instance_pt_x=instance_pt[0,0]
instance_pt_y=instance_pt[0,1]
plt.figure(figsize=(10,10))
plt.subplot(1,2,1)
plt.imshow(instance_pt_x)
plt.title('instance_pt_x')
plt.subplot(1,2,2)
plt.imshow(instance_pt_y)
plt.title('instance_pt_y')
```

```python
height_pt=test_cnnseg.outblobs['height_pt']
plt.figure(figsize=(5,5))
print(height_pt.shape)
height_pt=height_pt[0,0]
plt.imshow(height_pt)
plt.title('height_pt')
plt.colorbar()
```

```python
class_score=test_cnnseg.outblobs['class_score']
print(class_score.shape)
plt.figure(figsize=(20,12))
plt.subplot(2,3,1)
plt.imshow(class_score[0,0])
plt.colorbar()
plt.subplot(2,3,2)
plt.imshow(class_score[0,1])
plt.colorbar()
plt.subplot(2,3,3)
plt.imshow(class_score[0,2])
plt.colorbar()
plt.subplot(2,3,4)
plt.imshow(class_score[0,3])
plt.colorbar()
plt.subplot(2,3,5)
plt.imshow(class_score[0,4])
plt.colorbar()
```

```python
category_score=test_cnnseg.outblobs['category_score']
print(category_score.shape)
plt.figure(figsize=(5,5))
plt.imshow(category_score[0,0])
```

```python
confidence_score=test_cnnseg.outblobs['confidence_score']
print(confidence_score.shape)
plt.figure(figsize=(5,5))
plt.imshow(confidence_score[0,0])
plt.colorbar()
```
